```python
import ast
```

```python
def convert_features(feature):
    res = []
    for i in ast.literal_eval(feature):
        res.append(i['name'])
    return res
```

```python
movies['genres'] = movies['genres'].apply(convert_features)
```

```python
movies.iloc[0].keywords
```

```python
movies['keywords'] = movies['keywords'].apply(convert_features)
```

```python
movies.info()
```

```python
def convert_cast(cast):
    res = []
    count = 0
    for i in ast.literal_eval(cast):
        if count < 4:
            res.append(i['name'])
            count += 1
        else:
            break
    return res
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.info()
```

```python
def convert_crew(crew):
    res = []
    for i in ast.literal_eval(crew):
        if i['job'] == 'Director':
            res.append(i['name'])
            break
            
    return res
```

```python
movies['crew'] = movies['crew'].apply(convert_crew)
```

```python
movies.iloc[0].crew
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
movies[movies['overview'].apply(len) == 0]
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres']*3 + movies['keywords']*2 + movies['cast']*2 + movies['crew']*3
```

```python
df = movies[['id', 'title', 'tags']]
```

```python
df['tags'] = df['tags'].apply(lambda x: " ".join(x))
```

```python
df['tags'] = df['tags'].apply(lambda x: x.lower())
```

```python
df.head()
```

```python
df['tags'][1]
```

## Vecterization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.stem.porter import PorterStemmer
```

```python
Tfidf = TfidfVectorizer(max_features=5000, stop_words='english')
ps = PorterStemmer()
```

```python
def stem(text):
    res = []
    for i in text.split():
        res.append(ps.stem(i))
    return " ".join(res)
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
vectors = Tfidf.fit_transform(df['tags'])
```

## Model Traning

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = df[df['title'] == movie].index[0]
    target = similarity[movie_index]
    movies_list = sorted(list(enumerate(target)), reverse=True, key=lambda x: x[1])[1:6]
    
    for i in movies_list:
        print(df.iloc[i[0]].title)
```

```python
recommend('Iron Man')
```
